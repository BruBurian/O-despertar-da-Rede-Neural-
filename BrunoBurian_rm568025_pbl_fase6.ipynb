```python
import os
import shutil

base_path = "/content/drive/MyDrive/Colab Notebooks"
dataset_path = f"{base_path}/dataset"
cnn_path = f"{base_path}/cnn_dataset"

# Criar estrutura
for split in ["train", "val"]:
    for classe in ["banana", "microfone"]:
        os.makedirs(f"{cnn_path}/{split}/{classe}", exist_ok=True)

# Função para classificar imagem
def get_class(filename):
    filename = filename.lower()
    if filename.startswith("b"):
        return "banana"
    elif filename.startswith("m"):
        return "microfone"
    else:
        return None

# Copiar arquivos
for split in ["train", "val"]:
    img_folder = f"{dataset_path}/{split}/images"

    for file in os.listdir(img_folder):
        classe = get_class(file)

        if classe:
            src = os.path.join(img_folder, file)
            dst = os.path.join(cnn_path, split, classe, file)
            shutil.copy(src, dst)

print("✅ Dataset da CNN criado com sucesso!")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

img_size = 224
batch_size = 8

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "/content/drive/MyDrive/Colab Notebooks/cnn_dataset/train",
    image_size=(img_size, img_size),
    batch_size=batch_size
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "/content/drive/MyDrive/Colab Notebooks/cnn_dataset/val",
    image_size=(img_size, img_size),
    batch_size=batch_size
)

model = models.Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(32, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(2, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(train_ds, validation_data=val_ds, epochs=10)
```

###Análise da CNN

A rede neural convolucional (CNN) foi treinada do zero com o dataset construído, contendo imagens das classes “banana” e “microfone”.

Durante o treinamento, o modelo atingiu alta acurácia, chegando a valores próximos de 100% no conjunto de treino e entre 75% e 100% na validação.

Esses resultados indicam que o modelo foi capaz de aprender a diferenciar as classes propostas de forma eficiente.

No entanto, devido ao tamanho reduzido do dataset, há indícios de overfitting, uma vez que o modelo apresenta desempenho muito elevado no treino, podendo não generalizar bem para novos dados.

#Conclusão

Foram analisadas três abordagens distintas para o problema de visão computacional: YOLO customizado, YOLO pré-treinado e uma CNN treinada do zero.

O YOLO customizado demonstrou capacidade de adaptação ao problema, sendo capaz de detectar os objetos específicos definidos no dataset, ainda que com limitações de desempenho devido à baixa quantidade de dados.

Já a CNN apresentou os melhores resultados em termos de acurácia, sendo altamente eficiente na tarefa de classificação. No entanto, sua limitação está no fato de não realizar detecção de objetos, apenas classificação da imagem como um todo.

Dessa forma, conclui-se que a escolha da abordagem depende diretamente do problema a ser resolvido, sendo o YOLO mais adequado para detecção e a CNN para classificação.
